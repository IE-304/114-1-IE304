```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Poisson distribution parameter
lam = 5.6   #Lambda
```

```python
# Mean and std
mu = lam
sigma = np.sqrt(lam)
```

```python
# Dynamic range: mean ± 4σ, clipped at 0
lower = max(0, int(mu - 4 * sigma))
upper = int(mu + 4 * sigma)
```

```python
# Support of x values
x = np.arange(lower, upper + 1)
```

```python
# PMF values
pmf = poisson.pmf(x, lam)
```

```python
# Plot
FS=16
plt.figure(figsize=(6, 4))
plt.bar(x, pmf, color="steelblue", alpha=0.8, edgecolor="black")
#plt.title('Poisson Probability Mass Function (PMF)', fontsize=FS)
plt.title(label = 'Poisson Probability Mass Function (PMF),\n λ = '+str(lam), fontsize=FS, loc='center', pad=5)
textstr = f"{lam:.4f}: {mu:.4f}"
plt.xlabel('x', fontsize=FS-2)
plt.ylabel('p(x)', fontsize=FS-2)
plt.xticks(fontsize=FS-2)
plt.yticks(fontsize=FS-2)
plt.show()
```
